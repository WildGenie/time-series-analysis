```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d 
```

```python
df = pd.read_csv('wind_data.csv')
```

```python
df['ps'] = df['p'].shift(-1)


df.sort_values(['Wd1'], inplace=True)
df = df.dropna()
```

```python
df
```

```python
import trimesh as tri

data = df[['Wd1','Ws1','ps']].to_numpy()
print(data.shape)
mesh = tri.points.PointCloud(data).convex_hull
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(mesh.vertices[:,0], mesh.vertices[:,1], triangles=mesh.faces, Z=mesh.vertices[:,2]) 
```

```python
fig = plt.figure()

plt.hist(np.around(data[:,1], 2)
```

```python
np.around(np.unique(data[:,1]), 1)
```

```python
d_data = np.zeros((360, 360))

for row in data:
    d = int(np.floor(row[0]).astype(int))
    v = int(np.floor(row[1]*10).astype(int))
    #print(int(d))
    #print(type(v))
    d_data[d,v] = max(row[2], d_data[d,v])
```

```python
d_data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

d = np.arange(0,360,1)
v = np.arange(0,360,1)

ds, vs = np.meshgrid(d, v)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(ds.T, vs.T, d_data, rstride=1, cstride=1, cmap='viridis')
plt.show()
```

```python
fig = plt.figure()
plt.imshow(d_data.T/np.max(d_data),interpolation='nearest')
```
